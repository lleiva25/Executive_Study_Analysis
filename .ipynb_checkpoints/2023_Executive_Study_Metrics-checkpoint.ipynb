<h2> Clean up Dataset </h2>

```python

#Creating new dataframe based on relevant columns
Vallania_Data1 = pd.DataFrame(Vallania_RawData_Blood,columns=[
    "Created on",
    "Original Content (cntn_id)",
    "Type (cntp_name)",
    "Id",
    "Status",
    "Subject ID",
    "FlashLink Present",
    "FlashLink Alarm Triggered",
    "FlashLink Serial Number",
    "Temp Sensor Present",
    "Temp Sensor Trip Duration",
    "Temp Sensor Triggered Mins",
    "Nanocool Kit Present",
    "Nanocool Kit Activated"])

Vallania_Data2 = pd.DataFrame(Vallania_RawData_Accession,columns=[
    "Created on",
    "Type (cntp_name)",
    "Original Content (cntn_id)",
    "Id",
    "Status",
    "Subject ID",
    "Verified",
    ])

#Convert dataset to timeframe format
Vallania_Data1["Created on"]= pd.to_datetime(Vallania_Data1["Created on"], yearfirst=True, infer_datetime_format = True)
Vallania_Data1["Created on"]= Vallania_Data1["Created on"].dt.strftime('%Y/%m/%d')

Vallania_Data2["Created on"]= pd.to_datetime(Vallania_Data2["Created on"], yearfirst=True, infer_datetime_format = True)
Vallania_Data2["Created on"]= Vallania_Data2["Created on"].dt.strftime('%Y/%m/%d')

#Renaming Columns
Vallania_Data1 = Vallania_Data1.rename(columns={Vallania_Data1.columns[2]:"BCT IDs",
                              Vallania_Data1.columns[3]:"BCT Status",
                              Vallania_Data1.columns[4]:"BCT Subject ID"})
Vallania_Data2 = Vallania_Data2.rename(columns={Vallania_Data2.columns[2]:"Kit ID",
                                                Vallania_Data2.columns[3]:"Original Content (cntn_id)",
                                                Vallania_Data2.columns[4]:"Accession Status",
                                                Vallania_Data2.columns[5]:"Accession Subject ID"})

#Merging data bases
#Vallania_Data = pd.concat([Vallania_Data1, Vallania_Data2])
Vallania_Data = pd.merge(Vallania_Data1, Vallania_Data2, on='Original Content (cntn_id)')

#Shows the first 5 rows of the dataframe
Vallania_Data.head()

#Fill NaN with 0
Vallania_Data.fillna(0,inplace=True)

#See the new Dataframe
Vallania_Data.to_csv("Archive/2023_Vallania_Date_Merged.csv")
Vallania_Data.info()
```

```python
#Import calender module to align days with day of the week
import calendar

#Create a copy in order to keep the original dataset for other graphs
Vallania_Date_Daily = Vallania_Date.copy()
```

```python
#------------------------------------------------------
#Create bins for the quarter values
#Note: quarter=[0,3,6,9,12]
#Note: quarter_labels = ["Q1","Q2","Q3","Q4"]
#-------------------------------------------------------

#Creating a csv file for in order to create a new dataframe
Vallania_Date_Daily_csv = Vallania_Date_Daily.to_csv("Archive/Vallania_Date_Daily.csv")

#Having Jupyter read the newly created csv as a dataframe
Vallania_Date_Daily_Read = pd.concat(map(pd.read_csv, ["Archive/Vallania_Date_Daily.csv"]))

#Bin by Quarter by Month & Cleanup dataframe
Vallania_Date_Daily_Read["Month"] = pd.cut(Vallania_Date_Daily_Read["Month"], bins=quarter, labels= quarter_labels)
del Vallania_Date_Daily_Read["Unnamed: 0"]
del Vallania_Date_Daily_Read["Year"]
del Vallania_Date_Daily_Read["Day"]

Vallania_Date_Daily_Read
```

```python
Vallania_Date_Daily_Group = Vallania_Date_Daily_Read.groupby(["Month","Day of Week"]).agg(['count'])
del Vallania_Date_Daily_Group["Date"]
Vallania_Date_Daily_Group
```

```python
#Filter out the table by quarter
Vallania_Date_Daily_Q1 = Vallania_Date_Daily_Read[(Vallania_Date_Daily_Read["Month"]=="Q1")]
Vallania_Date_Daily_Q2 = Vallania_Date_Daily_Read[(Vallania_Date_Daily_Read["Month"]=="Q2")]
Vallania_Date_Daily_Q3 = Vallania_Date_Daily_Read[(Vallania_Date_Daily_Read["Month"]=="Q3")]
Vallania_Date_Daily_Q4 = Vallania_Date_Daily_Read[(Vallania_Date_Daily_Read["Month"]=="Q4")]

#Obtaining the Start & End Dates for each Quarter
Q1_Start = Vallania_Date_Daily_Q1['Date'].min()
Q1_End = Vallania_Date_Daily_Q1['Date'].max()

Q2_Start = Vallania_Date_Daily_Q2['Date'].min()
Q2_End = Vallania_Date_Daily_Q2['Date'].max()

Q3_Start = Vallania_Date_Daily_Q3['Date'].min()
Q3_End = Vallania_Date_Daily_Q3['Date'].max()

Q4_Start = Vallania_Date_Daily_Q4['Date'].min()
Q4_End = Vallania_Date_Daily_Q4['Date'].max()


#Import datetime module for function
import datetime

#Dictionary of items
Q1_Weekdays = {}
Q2_Weekdays = {}
Q3_Weekdays = {}
Q4_Weekdays = {}

#Create a function that will tell you the number of week days in quarter
def weekday_count(start, end):
  start_date  = datetime.datetime.strptime(start, '%Y-%m-%d')
  end_date    = datetime.datetime.strptime(end, '%Y-%m-%d')
  week        = {}
  for i in range((end_date - start_date).days):
    day       = calendar.day_name[(start_date + datetime.timedelta(days=i+1)).weekday()]
    week[day] = week[day] + 1 if day in week else 1
  return week

#Add function output to dictionary
Q1_Weekdays = weekday_count(Q1_Start,Q1_End)
Q2_Weekdays = weekday_count(Q2_Start,Q2_End)
Q3_Weekdays = weekday_count(Q3_Start,Q3_End)
Q4_Weekdays = weekday_count(Q4_Start,Q4_End)


#Find the total amount of samples per week day
Vallania_Date_Daily_Q1 = Vallania_Date_Daily_Q1.groupby("Day of Week").count()
Vallania_Date_Daily_Q2 = Vallania_Date_Daily_Q2.groupby("Day of Week").count()
Vallania_Date_Daily_Q3 = Vallania_Date_Daily_Q3.groupby("Day of Week").count()
Vallania_Date_Daily_Q4 = Vallania_Date_Daily_Q4.groupby("Day of Week").count()

#Using iloc to select individual rows, we are able to divded the total of samples by the number of days in week
#Repeat for every quarter
Vallania_Date_Daily_Q1['Month'].iloc[0] = Vallania_Date_Daily_Q1["Kit ID"].iloc[0]/(Q1_Weekdays.get('Monday'))
Vallania_Date_Daily_Q1['Month'].iloc[1] = Vallania_Date_Daily_Q1["Kit ID"].iloc[1]/(Q1_Weekdays.get('Tuesday'))
Vallania_Date_Daily_Q1['Month'].iloc[2] = Vallania_Date_Daily_Q1["Kit ID"].iloc[2]/(Q1_Weekdays.get('Wednesday'))
Vallania_Date_Daily_Q1['Month'].iloc[3] = Vallania_Date_Daily_Q1["Kit ID"].iloc[3]/(Q1_Weekdays.get('Thursday'))
Vallania_Date_Daily_Q1['Month'].iloc[4] = Vallania_Date_Daily_Q1["Kit ID"].iloc[4]/(Q1_Weekdays.get('Friday'))
Vallania_Date_Daily_Q1['Month'].iloc[5] = Vallania_Date_Daily_Q1["Kit ID"].iloc[5]/(Q1_Weekdays.get('Sunday'))

Vallania_Date_Daily_Q2['Month'].iloc[0] = Vallania_Date_Daily_Q2["Kit ID"].iloc[0]/(Q2_Weekdays.get('Monday'))
Vallania_Date_Daily_Q2['Month'].iloc[1] = Vallania_Date_Daily_Q2["Kit ID"].iloc[1]/(Q2_Weekdays.get('Tuesday'))
Vallania_Date_Daily_Q2['Month'].iloc[2] = Vallania_Date_Daily_Q2["Kit ID"].iloc[2]/(Q2_Weekdays.get('Wednesday'))
Vallania_Date_Daily_Q2['Month'].iloc[3] = Vallania_Date_Daily_Q2["Kit ID"].iloc[3]/(Q2_Weekdays.get('Thursday'))
Vallania_Date_Daily_Q2['Month'].iloc[4] = Vallania_Date_Daily_Q2["Kit ID"].iloc[4]/(Q2_Weekdays.get('Friday'))

Vallania_Date_Daily_Q3['Month'].iloc[0] = Vallania_Date_Daily_Q3["Kit ID"].iloc[0]/(Q3_Weekdays.get('Monday'))
Vallania_Date_Daily_Q3['Month'].iloc[1] = Vallania_Date_Daily_Q3["Kit ID"].iloc[1]/(Q3_Weekdays.get('Tuesday'))
Vallania_Date_Daily_Q3['Month'].iloc[2] = Vallania_Date_Daily_Q3["Kit ID"].iloc[2]/(Q3_Weekdays.get('Wednesday'))
Vallania_Date_Daily_Q3['Month'].iloc[3] = Vallania_Date_Daily_Q3["Kit ID"].iloc[3]/(Q3_Weekdays.get('Thursday'))
Vallania_Date_Daily_Q3['Month'].iloc[4] = Vallania_Date_Daily_Q3["Kit ID"].iloc[4]/(Q3_Weekdays.get('Friday'))

Vallania_Date_Daily_Q4['Month'].iloc[0] = Vallania_Date_Daily_Q4["Kit ID"].iloc[0]/(Q4_Weekdays.get('Monday'))
Vallania_Date_Daily_Q4['Month'].iloc[1] = Vallania_Date_Daily_Q4["Kit ID"].iloc[1]/(Q4_Weekdays.get('Tuesday'))
Vallania_Date_Daily_Q4['Month'].iloc[2] = Vallania_Date_Daily_Q4["Kit ID"].iloc[2]/(Q4_Weekdays.get('Wednesday'))
Vallania_Date_Daily_Q4['Month'].iloc[3] = Vallania_Date_Daily_Q4["Kit ID"].iloc[3]/(Q4_Weekdays.get('Thursday'))
Vallania_Date_Daily_Q4['Month'].iloc[4] = Vallania_Date_Daily_Q4["Kit ID"].iloc[4]/(Q4_Weekdays.get('Friday'))

#Reformat and clean table
Vallania_Date_Daily_Mean_Q1 = pd.DataFrame(Vallania_Date_Daily_Q1, columns=["Month"])
Vallania_Date_Daily_Mean_Q1 = Vallania_Date_Daily_Mean_Q1.round(1)
Vallania_Date_Daily_Mean_Q1 = Vallania_Date_Daily_Mean_Q1.rename(columns={'Month':'Q1'})
Vallania_Date_Daily_Mean_Q1      

Vallania_Date_Daily_Mean_Q2 = pd.DataFrame(Vallania_Date_Daily_Q2, columns=["Month"])
Vallania_Date_Daily_Mean_Q2 = Vallania_Date_Daily_Mean_Q2.round(1)
Vallania_Date_Daily_Mean_Q2 = Vallania_Date_Daily_Mean_Q2.rename(columns={'Month':'Q2'})
Vallania_Date_Daily_Mean_Q2 

Vallania_Date_Daily_Mean_Q3 = pd.DataFrame(Vallania_Date_Daily_Q3, columns=["Month"])
Vallania_Date_Daily_Mean_Q3 = Vallania_Date_Daily_Mean_Q3.round(1)
Vallania_Date_Daily_Mean_Q3 = Vallania_Date_Daily_Mean_Q3.rename(columns={'Month':'Q3'})
Vallania_Date_Daily_Mean_Q3 

Vallania_Date_Daily_Mean_Q4 = pd.DataFrame(Vallania_Date_Daily_Q4, columns=["Month"])
Vallania_Date_Daily_Mean_Q4 = Vallania_Date_Daily_Mean_Q4.round(1)
Vallania_Date_Daily_Mean_Q4 = Vallania_Date_Daily_Mean_Q4.rename(columns={'Month':'Q4'})

Vallania_Date_Daily_Mean_Q3 
```

<h2>Salesforce Data Cleanup</h2>

```python
#Find all Issue Types
print(Vallania_RawData_Deviation['Type of Issue'].unique())
```

```python
#Change Subject ID Column to string in order to split
Vallania_RawData_Deviation["Subject ID"] = Vallania_RawData_Deviation["Subject ID"].astype('str') # which will by default set the length to the max len it encounters

#Rename Columns
Vallania_RawData_Deviation = Vallania_RawData_Deviation.rename(columns={"Case Last Modified Date":"Closed Date"})  

#Recategorize Type of Issues to these categories 
#Missing Information, Mismatch Information, EDC/SLIMS Discrepancy, Duplicate Subject ID, Clotted, Hemolyzed, One/No Viable BCT, GDP/Smeared BCTs & Operator Error
Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['Missing Info on RF',
                                                                                                'All BCTs Blank',
                                                                                                'Missing Subj ID on BCTs',
                                                                                                'Missing DOC on BCTs',
                                                                                                'Missing Info on RF; Other'
                                                                                                ],'Missing Information')

Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['Sub ID Mismatch b/w req and bcts',
                                                                                                'Access ID Mismatch b/w req and bcts',
                                                                                                'Sub ID Mismatch w/in BCTs',
                                                                                                'DOC Mismatch b/w RF and BCTs'
                                                                                                ],'Mismatch Information')

Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['DOC Mismatch b/w RF and BCTs; EDC/SLIMS Discrepancy'
                                                                                                ],'EDC/SLIMS Discrepancy')

Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['Clotted',
                                                                                                 'Clotted; Only one viable BCT'
                                                                                                ],'Clotted')

Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['Collection Date > 7 days',
                                                                                                'Gross Hemolysis',
                                                                                                'Gross Hemolysis; Only one viable BCT'
                                                                                                ],'Hemolyzed/>7 DOC')

Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['Low volume',
                                                                                                'Only one viable BCT'
                                                                                                ],'One/No Viable BCT')

Vallania_RawData_Deviation['Type of Issue']=Vallania_RawData_Deviation['Type of Issue'].replace(['Other',
                                                                                                'Bad GDP',
                                                                                                'Smeared BCTs',
                                                                                                'No Req Form'
                                                                                                ],'Other')

#Confirming dtype change
Vallania_RawData_Deviation.iloc[0]
```

<h2>Average Ticket Resolution Time</h2>

```python
#Convert dataset to timeframe format in order to subtract columns
Vallania_Deviation_db['Resolution Time'] = (pd.to_datetime(Vallania_Deviation_db['Closed Date'])).sub(pd.to_datetime(Vallania_Deviation_db['Opened Date'])).dt.days

#Convert dtype to int
Vallania_Deviation_db['Resolution Time'] = Vallania_Deviation_db['Resolution Time'].astype(int)
Vallania_Deviation_db.head()
```

```python
#Create a copy of the dataframe to continue editing it
resolution_time_db = Vallania_Deviation_db.copy()

#Delete unnecessary columns
del resolution_time_db['Opened Date']
del resolution_time_db['Closed Date']

#Change column type to int
resolution_time_db["Resolution Time"] = resolution_time_db["Resolution Time"].astype(int)

#Change the negative values to zero
resolution_time_db["Resolution Time"].loc[resolution_time_db["Resolution Time"]<0] = 0

#Change the values that are over 14 days to 15 days
resolution_time_db["Resolution Time"].loc[resolution_time_db["Resolution Time"]>15] = 15

resolution_time_db.head()
```

```python
#Create a list of unique site IDs
Site_list = resolution_time_db['Site ID'].unique()

#Sort site IDS in ascending order
Site_list.sort()

Site_list
```

```python
Avg_resolution_time = []

for site in Site_list:
    x = resolution_time_db.loc[resolution_time_db['Site ID'] == site].mean(numeric_only=True)['Resolution Time']
    Avg_resolution_time.append(x)

Avg_resolution_time
```

```python
#Function for displaying a gradient come one side of the graph to the other
def hex_to_RGB(hex_str):
    """ #FFFFFF -> [255,255,255]"""
    #Pass 16 to the integer function for change of base
    return [int(hex_str[i:i+2], 16) for i in range(1,6,2)]

def get_color_gradient(c1, c2, n):
    """
    Given two hex colors, returns a color gradient
    with n colors.
    """
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]
```

```python
#Turning paired list data into dataframe
Avg_resolution_time_db = pd.DataFrame({
    'Site ID':Site_list,
    'Avg Resolution Time':Avg_resolution_time
})

#Sorting avg resolution time
Sort_avg_resolution_time_db = Avg_resolution_time_db.sort_values(by=Avg_resolution_time_db.columns[1],ascending=True)

#Create empty list. Obtain sorted list of site ID's through loop
sorted_site_ids = []

for i in Sort_avg_resolution_time_db['Site ID'].iloc[:len(Sort_avg_resolution_time_db)]:
    sorted_site_ids.append(i)

#Declaring colors for graph gradient
color1 = "#8A5AC2"
color2 = "#0000ff"


#Ploting dataframe
ax = Sort_avg_resolution_time_db.plot.bar(figsize=(20,15),
                                          width=0.9, 
                                          color = get_color_gradient(color1, color2, len(Sort_avg_resolution_time_db)),
                                          fontsize=15,
                                          align='center')

#Labeling x y axis
ax.set_ylabel('Average Resolution Time (Days)',fontsize=20)
ax.set_xlabel('Site ID',fontsize=20)

#Relabeling x_axis ticks
ax.set_xticklabels(sorted_site_ids, rotation=45)

#Labeling the bars in the graph
for c in ax.containers:
    ax.bar_label(c, fmt=lambda x: f'{x:.0f}' if x>0 else'', label_type='edge',fontsize=15)

ax.figure.savefig('2023 Data Plots/Salesforce_01_Stacked_Deviation_Bar_Graph.png')

plt.show()
```

<h2>Ticket Creation Rate by Site</h2>

```python
#Copy dataset for analysis
creation_rate_db = Vallania_Deviation_db.copy()

#Reset index
Vallania_Sites_reset = Vallania_Sites.reset_index(col_level=1)

Vallania_Sites_reset.head()
```

```python
#Get list of site IDs
list_sites = creation_rate_db['Site ID'].unique()

#Create empty list to save data into
num_ticket_list = []
total_num_samples = []

#Loop through column so that it counts the number of times a site filed a ticket
for site in list_sites:
    x = creation_rate_db.loc[creation_rate_db['Site ID'] == site].count(numeric_only=None)['Site ID']
    num_ticket_list.append(x)

for site in list_sites:
    x = Vallania_Sites_reset.loc[Vallania_Sites_reset[0] == site].sum(numeric_only=None)[1]
    total_num_samples.append(x)

#Create a dataframebased on newly created lists
rate_db = pd.DataFrame({
    'Site ID': list_sites,
    'Number of Deviations': num_ticket_list,
    'Total Samples From Site' : total_num_samples
})

rate_db.head()
```

```python
#Create new column for percentage rate
rate_db['Percentage Rate'] = round(100*(rate_db['Number of Deviations']/rate_db['Total Samples From Site']),1)

#Deleting unnecessary columns
del rate_db['Number of Deviations']
del rate_db['Total Samples From Site']

#Sort values by percentage rate
rate_sort_db = rate_db.sort_values(by=rate_db.columns[1],ascending=True)

rate_sort_db.head()
```

```python
ax = rate_sort_db.plot.barh(stacked = True,
                       figsize=(20,15),
                       width=0.8,
                       colormap='Set3',
                       fontsize=15,
                       align='center')

#Labeling x y axis
ax.set_ylabel('Site ID',fontsize=15)
ax.set_xlabel('Percentage',fontsize=15)

#Title graph
plt.title(f"Ticket Creation Rate Based on Deviations Per Site",fontsize=20)

#Relabing y axis ticks
ax.set_yticklabels(rate_sort_db['Site ID'].unique())

#Renaming the Legend
ax.legend(fontsize=20)

#Labeling the bars in the graph
for c in ax.containers:
    ax.bar_label(c, fmt=lambda x: f'{x:.0f}%' if x>0 else'', label_type='center',fontsize=15)
```
